```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/yelderiny/Projects/Dissertation/Data/project-data3.csv')
df.head()
```

```python
df.boxplot(column=['size', 'age', 'contributor_xp1', 'contributor_xp2', 'contributor_xp3', 'pull_requests', 'contributors'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["size"].mean() - (3 * df["size"].std())
upper_limit = df["size"].mean() + (3 * df["size"].std())

df["size"] = np.where(df["size"]> upper_limit, upper_limit, np.where(df["size"]< lower_limit, lower_limit,df["size"]))

df.boxplot(column=['size'], figsize=(10,7), rot=45)
plt.show()
```
