```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/yelderiny/Projects/Dissertation/Data/project-data.csv')
df.head()
```

```python
df.boxplot(column=['pull_requests', 'size', 'contributors', 'age', 'contributor_xp', 'pr_points'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["size"].mean() - (3 * df["size"].std())
upper_limit = df["size"].mean() + (3 * df["size"].std())

size_outliers = np.where(df["size"]> upper_limit, True,
                np.where(df["size"]< lower_limit, True, False))

df = df.loc[~size_outliers]

df.boxplot(column=['size'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["contributor_xp"].mean() - (3 * df["contributor_xp"].std())
upper_limit = df["contributor_xp"].mean() + (3 * df["contributor_xp"].std())

contributor_xp_outliers = np.where(df["contributor_xp"]> upper_limit, True,
                np.where(df["contributor_xp"]< lower_limit, True, False))

df = df.loc[~contributor_xp_outliers]

df.boxplot(column=['contributor_xp'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["pull_requests"].mean() - (3 * df["pull_requests"].std())
upper_limit = df["pull_requests"].mean() + (3 * df["pull_requests"].std())

pull_requests_outliers = np.where(df["pull_requests"]> upper_limit, True,
                np.where(df["pull_requests"]< lower_limit, True, False))

df = df.loc[~pull_requests_outliers]

df.boxplot(column=['pull_requests'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["contributors"].mean() - (3 * df["contributors"].std())
upper_limit = df["contributors"].mean() + (3 * df["contributors"].std())

contributors_outliers = np.where(df["contributors"]> upper_limit, True,
                np.where(df["contributors"]< lower_limit, True, False))

df = df.loc[~contributors_outliers]

df.boxplot(column=['contributors'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["pr_points"].mean() - (3 * df["pr_points"].std())
upper_limit = df["pr_points"].mean() + (3 * df["pr_points"].std())

pr_points_outliers = np.where(df["pr_points"]> upper_limit, True,
                np.where(df["pr_points"]< lower_limit, True, False))

df = df.loc[~pr_points_outliers]

df.boxplot(column=['pr_points'], figsize=(10,7), rot=45)
plt.show()
```

```python
df.boxplot(column=['pull_requests', 'size', 'contributors', 'age', 'contributor_xp', 'pr_points'], figsize=(10,7), rot=45)
plt.show()
```
