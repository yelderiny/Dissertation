```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression, ElasticNet, Lasso, Ridge
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df1 = pd.read_csv('/Users/yelderiny/Projects/Dissertation/Data/processed-data.csv')
df1.head()
```

```python
df2 = pd.read_csv('/Users/yelderiny/Projects/Dissertation/Data/processed-data-outliers-capped.csv')
df2.head()
```

```python
features1 = df1.drop('pr_points', axis=1)
target1 = df1['pr_points']
```

```python
features2 = df2.drop('pr_points', axis=1)
target2 = df2['pr_points']
```

```python
max_error_scoring = "max_error"
neg_mean_absolute_error_scoring = "neg_mean_absolute_error"
r2_scoring = "r2"
neg_mean_squared_error_scoring = "neg_mean_squared_error"
```

```python
models = [
    ('Linear Regression', LinearRegression()),
    ('LASSO', Lasso()),
    ('Ridge', Ridge()),
    ('Elastic Net', ElasticNet()),
    ('Random Forest', RandomForestRegressor())
]
```

```python
for name, model in models:
    kfold = KFold(n_splits=10, shuffle=True, random_state=299)
    cv_results1 = cross_val_score(model, features1, target1, cv=kfold, scoring=max_error_scoring)
    cv_results2 = cross_val_score(model, features1, target1, cv=kfold, scoring=neg_mean_absolute_error_scoring)
    cv_results3 = cross_val_score(model, features1, target1, cv=kfold, scoring=neg_mean_squared_error_scoring)
    cv_results4 = cross_val_score(model, features1, target1, cv=kfold, scoring=r2_scoring)
    msg = f"{name} \n max error: {cv_results1.mean()} \n mean absolute error: {-cv_results2.mean()} \n max squared error: {-cv_results3.mean()} \n r2: {cv_results4.mean()}"
    print(msg, '\n', '-' * 30)
```

```python
for name, model in models:
    kfold = KFold(n_splits=10, shuffle=True, random_state=299)
    cv_results1 = cross_val_score(model, features2, target2, cv=kfold, scoring=max_error_scoring)
    cv_results2 = cross_val_score(model, features2, target2, cv=kfold, scoring=neg_mean_absolute_error_scoring)
    cv_results3 = cross_val_score(model, features2, target2, cv=kfold, scoring=neg_mean_squared_error_scoring)
    cv_results4 = cross_val_score(model, features2, target2, cv=kfold, scoring=r2_scoring)
    msg = f"{name} \n max error: {cv_results1.mean()} \n mean absolute error: {-cv_results2.mean()} \n max squared error: {-cv_results3.mean()} \n r2: {cv_results4.mean()}"
    print(msg, '\n', '-' * 30)
```
