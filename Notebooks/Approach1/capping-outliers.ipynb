```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# df = pd.read_csv('/Users/yelderiny/Projects/Dissertation/Data/processed-data.csv')
df = pd.read_csv('/Users/yelderiny/Projects/Dissertation/Data/project-data.csv')
df.head()
```

```python
df.boxplot(column=['size', 'age', 'contributor_xp', 'pull_requests', 'contributors'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["size"].mean() - (3 * df["size"].std())
upper_limit = df["size"].mean() + (3 * df["size"].std())

df["size"] = np.where(df["size"]> upper_limit, upper_limit, np.where(df["size"]< lower_limit, lower_limit,df["size"]))

df.boxplot(column=['size'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["contributor_xp"].mean() - (3 * df["contributor_xp"].std())
upper_limit = df["contributor_xp"].mean() + (3 * df["contributor_xp"].std())

df["contributor_xp"] = np.where(df["contributor_xp"]> upper_limit, upper_limit, np.where(df["contributor_xp"]< lower_limit, lower_limit,df["contributor_xp"]))

df.boxplot(column=['contributor_xp'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["pull_requests"].mean() - (3 * df["pull_requests"].std())
upper_limit = df["pull_requests"].mean() + (3 * df["pull_requests"].std())

df["pull_requests"] = np.where(df["pull_requests"]> upper_limit, upper_limit, np.where(df["pull_requests"]< lower_limit, lower_limit,df["pull_requests"]))

df.boxplot(column=['pull_requests'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["contributors"].mean() - (3 * df["contributors"].std())
upper_limit = df["contributors"].mean() + (3 * df["contributors"].std())

df["contributors"] = np.where(df["contributors"]> upper_limit, upper_limit, np.where(df["contributors"]< lower_limit, lower_limit,df["contributors"]))

df.boxplot(column=['contributors'], figsize=(10,7), rot=45)
plt.show()
```

```python
lower_limit = df["pr_points"].mean() - (3 * df["pr_points"].std())
upper_limit = df["pr_points"].mean() + (3 * df["pr_points"].std())

df["pr_points"] = np.where(df["pr_points"]> upper_limit, upper_limit, np.where(df["pr_points"]< lower_limit, lower_limit,df["pr_points"]))

df.boxplot(column=['pr_points'], figsize=(10,7), rot=45)
plt.show()
```

```python
df.boxplot(column=['size', 'age', 'contributor_xp', 'pull_requests', 'contributors'], figsize=(10,7), rot=45)
plt.show()
```
